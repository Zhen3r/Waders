```python
import rasterio as rio
import matplotlib.pyplot as plt
import numpy as np
```

```python
with rio.open("./tif/MCD12Q1.006_LC_Type1_doy2010001_aid0001.tif","r")as f:
    print(f.meta)
    a = f.read(1)
    
with rio.open("./tif/MCD12Q1.006_LC_Type1_doy2020001_aid0001.tif","r")as f:
    print(f.meta)
    b = f.read(1)
```

```python
x,y,r = 1000,1600,1200
aa = a[x:x+r,y:y+r]
bb = b[x:x+r,y:y+r]
cc = aa-bb

f,ax = plt.subplots(1,2,dpi=75,figsize=(10,8))

i = ax[0].imshow(aa,vmin=1, vmax=21, cmap="tab20")
plt.colorbar(i,ax=ax[0],ticks=range(1,21))
ax[0].axis("off")

i = ax[1].imshow(bb,vmin=1, vmax=21, cmap="tab20")
f.colorbar(i,ax=ax[1],ticks=range(1,21))
ax[1].axis("off")


plt.show()
```
